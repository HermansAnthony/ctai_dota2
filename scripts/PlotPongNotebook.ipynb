```python
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
# fetch data
x_deepq = [89638, 308659, 504467 ,688803, 871049, 1000000]
y_deepq = [-20.4, -11.1, 15.4, 16.7, 17.4, 19.1]

ppo = pd.read_csv('experiments\logs\Pong\ppo\progress.csv')
x_ppo = ppo['total_timesteps']
y_ppo = ppo['eprewmean']

trpo = pd.read_csv('experiments\logs\Pong\\trpo\progress.csv')
x_trpo = trpo['TimestepsSoFar']
y_trpo = trpo['EpRewMean']
```

```python
# Plot the results
plt.xlabel('time_steps')
plt.ylabel('reward')
plt.xlim(0, 1000000)

plt.plot(x_deepq, y_deepq, label='deepq', c='green')
plt.plot(x_ppo, y_ppo, label='ppo', c='red')
plt.plot(x_trpo, y_trpo, label='trpo', c='blue')

    
plt.legend(loc='best',  bbox_to_anchor=(0.90, 0.50, 0.5, 0.5))
plt.show()
```
